```python
import pandas as pd
import pandas as numpy
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel("PADDY DATA SET.xlsx")
df.head()
```

```python
df.info()
```

```python
df.dropna(inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
df=df.apply(LabelEncoder().fit_transform)
df
```

```python
X=df.iloc[:,0:18]
y=df.iloc[:,-1]
```

```python
features_name=tuple(X.columns)
features_name
```

```python
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
from sklearn.neighbors import KNeighborsClassifier as knn
from sklearn.linear_model import LogisticRegression as LGR
from sklearn.ensemble import RandomForestClassifier as rfc
```

```python
sbs = SFS(#knn(n_neighbors=3),
           #rfc(n_jobs=8),
           LGR(max_iter=1000),
           k_features='best', 
           forward=False, 
           floating=False, 
           verbose=2,
           #scoring = 'neg_mean_squared_error',  # sklearn regressors
           scoring='accuracy',  # sklearn classifiers
           cv=3)
sbs = sbs.fit(X, y,custom_feature_names=features_name)
```

```python
sbs.k_feature_names_,sbs.k_feature_idx_
```

```python
df = pd.DataFrame.from_dict(sbs.get_metric_dict()).T
df = df[["feature_idx","avg_score"]]
```

```python
from mlxtend.plotting import plot_sequential_feature_selection as plot_sbs 
fig = plot_sbs(sbs.get_metric_dict(), kind='std_err')
print("Selected Features :", sbs.k_feature_names_)
print("Selected Features ID :", sbs.k_feature_idx_)
plt.title('Sequential Forward Selection (w. StdErr)')
plt.grid()
plt.show()
```
