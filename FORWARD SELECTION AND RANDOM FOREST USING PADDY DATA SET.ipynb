```python
import pandas as pd
```

```python
from mlxtend.feature_selection import SequentialFeatureSelector
```

```python
df=pd.read_excel("PADDY DATA SET.xlsx")
df.head()
```

```python
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder
df=df.apply(LabelEncoder().fit_transform)
df
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df[df.isnull().any(axis=1)]
```

```python
x=df.iloc[:,0:18]
y=df.iloc[:,-1]
```

```python
x.head()
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=100)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
x_train.shape
```

```python
forward_feature_selection=SequentialFeatureSelector(RandomForestClassifier(n_jobs=-1),k_features=18,forward=True,floating=False,verbose=2,scoring="accuracy",cv=5).fit(x_train,y_train)
```

```python
forward_feature_selection.k_feature_idx_
```

```python
forward_feature_selection.k_feature_names_
```

```python
forward_feature_selection.k_score_
```
